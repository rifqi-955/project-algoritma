Algoritma Euclidean Distance

Rumus: $d = \sqrt{\sum (x_i - y_i)^2}$

Pseudocode:

```
Function EuclideanDistance(x1, x2, n):
  sum = 0
  
  For i from 1 to n:
    dif = x1[i] - x2[i]
    sum = sum + (dif ** 2)

  end For

  distance = sqrt(sum)
  return distance

end Function
```

dimana x1 dan x2 adalah koordinat array

n adalah jumlah dimensi

````python
import math

def euclidean(x1, x2, n):
  sum = 0
  for i in range(n):
    dif = x1[i] - x2[i]
    sum += (dif ** 2)
  distance = math.sqrt(sum)
  return distance

point1 = [1, 2]
point2 = [4, 5]
dimension = 2

print(euclidean(point1, point2, dimension))
````
