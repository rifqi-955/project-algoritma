Algoritma Selection Short
Kompleksitas Waktu:

Best case = $O(n^2)$

Average case = $O(n^2)$

Worst case = $O(n^2)$


Pseudocode:

```
 procedure selectionSort(A)
    n = length(A)
    for i = 0 to n - 2 do
        minIndex = i
        for j = i + 1 to n - 1 do
            if A[j] < A[minIndex] then
                minIndex = j
            end if
        end for
            swap(A[i], A[minIndex])
        end if
    end for
end function
```



````python
# Python code

def selectionSort(arr):
  n = len(arr)
  for i in range(n - 1):
    min = i
    for j in range(i + i, n):
      if arr[j] < arr[min]:
        min = j

    arr[i], arr[min] = arr[min], arr[i]

  return arr

sample = [87, 23, 56, 11, 43, 67]
print(selectionSort(sample))
````

Algoritma Bubble Sort
Kompleksitas Waktu:

Best case: $O(n)$

Average case: $O(n^2)$

Worst case: $O(n^2)$

Pseudocode:
```
BubbleSort(array):
  n = length(array)

  for i = 0 to n - 1 do
    swapped = false
    for j = 0 to n - i - 2:
      if array[j] > array[j + 1] then
        temp = array[j]
        array[j] = array[j + 1]
        array[j + 1] = temp
          swapped = TRUE
      end if
            
    end for
        
    if swapped == false then
      break
    end if
        
  end for
  
  return array
end function
```

````python
# Python code

def bubbleSort(arr):
  n = len(arr)
  for i in range(n):
    swapped = False
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        swapped = True
    if swapped == False:
      break
  return arr

sample = [64, 34, 23, 11, 90]
print(bubbleSort(sample))
````

Algoritma Insertion Sort
Kompleksitas Waktu:

Best case: $O(n)$

Average case: $O(n^2)$

Worst case: $O(n^2)$

```
Procedure InsertionSort(A)
    n = length(A)
    For i = 1 To n - 1 do
        key = A[i]
        
        j = i - 1
        
        While j >= 0 And A[j] > key Do
            A[j + 1] = A[j]
            j = j - 1
        End While
        
        A[j + 1] = key
        
    End For
End function
```

````python
# Python code

def insertionSort(arr):
  for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
      arr[j + 1] = arr[j]
      j -= 1
    arr[j + 1] = key

sample = [12, 11, 13, 5, 6]
print(insertionSort(sample))
````

Algoritma Counting Sort
Kompleksitas waktu:

Best case: $O(n + k)$

Average case: $O(n + k)$

Worst case: $O(n + k)$

Pseudocode:
```
Function CountSort(arr)
    n = len(arr)
    maxval = max(arr)

    cntArr = [0] * (maxval + 1)
    For Each v In arr Do
        cntArr[v] = cntArr[v] + 1
    End For

    For i = 1 To maxval Do
        cntArr[i] = cntArr[i] + cntArr[i - 1]
    End For

    ans = [0] * n
    
    For i = (n - 1) Down To 0 Do
        v ← arr[i]
        ans[cntArr[v] - 1] ← v
        cntArr[v] ← cntArr[v] - 1
    End For

    Return ans
END Function
```

````python
# Python code

def count_sort(arr):
    n = len(arr)
    maxval = max(arr)

    cntArr = [0] * (maxval + 1)

    for v in arr:
        cntArr[v] += 1

    for i in range(1, maxval + 1):
        cntArr[i] += cntArr[i - 1]

    ans = [0] * n
    for i in range(n - 1, -1, -1):
        v = arr[i]
        ans[cntArr[v] - 1] = v
        cntArr[v] -= 1

    return ans


sample = [2, 5, 3, 0, 2, 3, 0, 3]
print(count_sort(sample))
````
