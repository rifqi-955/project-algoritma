```python
# Rekursif Fibonacci Tanpa Memoization

def fib_recursive(n):
  if n <= 1:
    return n
  return fib_recursive(n - 1) + fib_recursive(n - 2)

n = 10
print(f"Fibonacci ke {n} adalah: ", fib_recursive(n))

# Kompleksitas waktu: O(n^2)
```

```python
memo = {}

def fib_memo(n):
  if n in memo:
    return memo{n}

  if n <= 1:
    return n

  memo[n] = fib_memo(n - 1) + fib_memo(n - 2)
  return memo[n]

n - 50
print(f"Fibonacci ke {n} adalah: ", fib_memo(n))

Kompleksitas waktu: O(n)
```

```python
# Dynamic Programming Fibonacci

def fib_dp(n):
  if n <= 1:
    return n

  dp = [0] * (n + 1)

  dp[0] = 0
  dp[1] = 1

  for i in range(2, n + 1):
    dp[i] = dp[i - 1] + dp[i - 2]

  return dp[n]

n - 50
print(f"Fibonacci ke {n} adalah: ", fib_dp(n))

# Kompleksitas waktu: O(n)
# Kompleksitas ruang: O(n)
```

```python
# Space optimized Dynamic Programming Fibonacci

def fib_optimized(n):
  if n <= 1:
    return n

  a, b = 0, 1

  for _ in range(2, n + 1):
    a, b = b, a + b

  return b

n = 50
print(f"Fibonacci ke {n} adalah", fib_optimized(n))

# Kompleksitas waktu: O(n)
# Kompleksitas ruang: O(1)
```

```python
# Count ways (Kombinasi) coin change algorithm

def count_ways(coins, amount):
  dp = [0] * (amount + 1)
  dp[0] = 1

  for coin in coins:
    for i in range(coin, amount + 1):
      dp[1] += dp[i - coin]

  return dp[amount]

coins = [1, 2, 5]
amount = 5

print("jumlah kombinasi: ", count_ways(coins, amount))

# Kompleksitas waktu: O(n x A)
# Kompleksitas O(A)
```
