```python
# Linear Search
def pencarian_linear(arr, target):
  for i in range(len(arr)):
    if arr[i] == target:
      return i
  return -1

# Contoh penggunaan
data = [10, 20, 30, 40, 50]
target = 30
print(pencarian_linear(data, target))
```

Algoritma ini mempunyai big O = O(n) dikarenakan terdapat for loop di baris ke-2 program

Analis case:

- Best case = O(1) --> algoritma langsung berhenti tepat hasil ditemukan di indeks pertama
- Average case = O(n) --> karena pertumbuhan waktu sebanding dengan n
- Worst case = O(n) --> algoritma harus melintasi seluruh elemen n dalam array

```python
# Binary search

def pencarian_binary(arr, target):
  kiri = 0
  kanan = len(arr) - 1

  while kiri <= kanan:
    tengah = (kiri + kanan) // 2

    if arr[tengah] == target:
      return tengah
    elif arr[tengah] < target:
      kiri = tengah + 1
    else:
      kanan = tengah - 1

  return -1

data = [10, 20, 30, 40, 50]
target = 40

print(pencarian_binary(data, target)) # Output: 3
```

Algoritma ini mempunyai big O = O(log n)

Analisis case:

- Best case: O(1) --> algoritma langsung return value ketika menemukan target ditengah
- Average case: O(log n) --> algoritma mencari value dengan waktu log2 n
- Worst case: O(log n) --> algoritma mencari value di pojok kanan atau kiri atau tidak ada

```python
import math

def jump_search(arr, target):
  n = len(arr)
  langkah = int(math.sqrt(n))
  prev = 0

  while arr[min(langkah, n) - 1] < target:
    prev = langkah
    langkah += int(math.sqrt(n))
    if prev >= n:
      return -1
    # linear search di block
    for i in range(prev, min(langkah, n)):
      if arr[i] == target:
        return i
  return -1

target = 60
data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, target)) # Output: 5
```

Algoritma ini mempunyai big O = O(√n)

Analisa case:

- Best case: O(1) --> algoritma hanya melakukan satu lompatan
- Average case: O(√n) --> algoritma memerlukan n/m lompatan dan sekitar m/2 perbandingan linear di dalam blok terakhir
- Worst case: O(√n) --> algoritma harus melakukan sebanyak n/m lompatan untuk mencapai blok terakhir

```python
# Interpolation search

def interpolation_search(arr, target):
  bawah = 0
  atas = len(arr) - 1

  while bawah <= atas and target >= arr[bawah] and target <= arr[atas]:
    if bawah == atas:
      if arr[bawah] == target:
        return bawah
      return -1

    # Rumus interpolasi
    pos = bawah + ((target - arr[bawah]) * (atas - bawah)) // (arr[atas] - arr[bawah])

    if arr[pos] == target:
      return pos
    elif arr[pos] < target:
      bawah = pos + 1
    else:
      atas = pos - 1

  return -1

data = [10, 20, 30, 40, 50, 60, 70]
target = 50

print(interpolation_search(data, target)) # Output: 4
```

Algoritma (Big O) dan analisa case:

- Best Case: O(1) --> Algoritma langsukan menemukan pada posisi pertama yang dihitung menggunakan rumus interpolasi.
- Average Case: O(log(log(n))) --> Berlaku jika data terdistribusi secara seragam (uniformly distributed). Algoritma ini memotong ruang pencarian secara eksponensial jauh lebih cepat dibandingkan binary search.
- Worst Case: O(n) --> Terjadi jika data terdistribusi secara tidak merata (misalnya, jika selisih nilai antar elemen sangat eksponensial, seperti 1, 2, 4, 8, 16). Pada kondisi ekstrem seperti ini, rumus posisi akan selalu bergeser satu langkah pada setiap iterasi, mirip dengan pencarian linear.
